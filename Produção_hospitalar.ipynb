```python
import pandas as pd
```

```python
dados = pd.read_csv("/content/A152918189_28_143_208.csv",
    encoding='ISO-8859-1',
    engine='python',  # Especifica o motor Python
    skiprows=3,       # Pula as 3 primeiras linhas
    sep=';',          # Usa o ponto e vírgula como separador
    skipfooter=12,    # Pula as 12 últimas linhas
    thousands='.',    # Considera o ponto como separador de milhares
    decimal=','       # Considera a vírgula como separador decimal
)
```
